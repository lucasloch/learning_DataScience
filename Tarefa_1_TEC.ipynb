```python
import pandas as pd
alunos = pd.DataFrame({'Nome': ['Ary', 'Cátia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
                        'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
                        'Idade': [15, 27, 56, 32, 42, 21, 19, 35], 
                        'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6]}, 
                        columns = ['Nome', 'Idade', 'Sexo', 'Notas'])

alunos
```

**Considerando o código da 14ª questão, faça o seguinte conjunto de operações de criação de novas variáveis: (i) Nova variável que indica a diferença entre cada nota e a média das notas, evidenciando o quanto cada aluno ficou acima ou abaixo da média geral; (ii) Nova variável que indica três faixas etárias (menor que 20 anos, entre 20 e 40 anos, maior que 40 anos).**

```python
# I - Nova variável que indica a diferença entre cada nota e a média das notas, 
# evidenciando o quanto cada aluno ficou acima ou abaixo da média geralalunos
alunos['Dif. Média Geral'] = (alunos['Notas'] - alunos['Notas'].mean()).round(1)
alunos
```

```python
import pandas as pd

dados = pd.read_csv('aluguel_amostra.csv', sep=';')

dados.head(n=10)
```

```python
dados
```

```python
# O quartil inferior (Q1) é a mediana da metade inferior dos dados
Q1 = dados['Valor m2'].quantile(0.25)
# O quartil superior (Q3) é a mediana da metade superior dos dados
Q3 = dados['Valor m2'].quantile(0.75)
# IIQ é o intervalo-interquartil, representado pela diferença dentre o terceiro e primeiro quartil
IIQ = (Q3 - Q1).round(2)
# Limite inferior
lim_inf = (Q1 - 1.5 * IIQ).round(2)
# Limite superior
lim_sup = (Q3 + 1.5 * IIQ).round(2)
# Mediana ou  Q2
Q2 = dados['Valor m2'].median()

data = [['Q1', Q1], ['Q2', Q2], ['Q3', Q3], ['Limite Inferior', lim_inf], ['Limite Superior', lim_sup]]

estatisticas = pd.DataFrame([Q1, Q2, Q3, lim_inf, lim_sup], index=['Q1', 'Q2', 'Q3', 'Limite Inferior', 'Limite Superior'], columns=['Estatísticas'])
estatisticas
```
